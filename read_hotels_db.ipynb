```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('french_hotels.db')

# Query the database and filter the data
query = "SELECT * FROM french_hotels WHERE temperature > 29 AND overall_rank > 8.5"
filtered_df = pd.read_sql_query(query, conn)

# Display the filtered data
print(filtered_df)

# Close the connection
conn.close()
```
